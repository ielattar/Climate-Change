```python
import missingno as msno
import numpy as np
%matplotlib inline
import pandas as pd
```

```python
all_data_filename = 'D:\\temp\\all.csv'
alldata_df = pd.read_csv(all_data_filename,engine='python').reset_index()
alldata_df.drop(alldata_df.columns[[1]], axis=1, inplace=True)
msno.matrix(alldata_df)


```
